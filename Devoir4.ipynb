```python
import networkx as nx
import numpy as np
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
```

```python
edges = [
    ('La mouquette','Officier',1),
    ('La mouquette','Catherine Maheu',3),
    ('La mouquette','Etienne Lantier',2),
    ('Maigrat','La maheude',2),
    ('Maigrat','Chaval',1),
    ('Maigrat','Catherine Maheu',1),
    ('Rasseneur','Etienne Lantier',1),
    ('Rasseneur','Chaval',1),
    ('Rasseneur','Toussaint Maheu',2),
    ('Bonemort','La maheude',2),
    ('Bonemort','Etienne Lantier',1),
    ('Bonemort','Cécile',3),
    ('Contre-maitre','Toussaint Maheu',1),
    ('Officier','La maheude',1),
    ('Officier','Toussaint Maheu',3),
    ('Officier','Etienne Lantier',1),
    ('Officier','Négrel',1),
    ('Catherine Maheu','La maheude',3),
    ('Catherine Maheu','Etienne Lantier',5),
    ('Catherine Maheu','Toussaint Maheu',3),
    ('Catherine Maheu','Chaval',5),
    ('Catherine Maheu','Cécile',2),
    ('Chaval','Etienne Lantier',5), 
    ('Chaval','Toussaint Maheu',3),
    ('Chaval','La maheude',2),
    ('Chaval','Deneulin',3),
    ('Chaval','Négrel',1),
    ('Les Grégoire','Cécile',3),
    ('Les Grégoire','Deneulin',1),
    ('Les Grégoire','Négrel',2),
    ('Les Grégoire','Philippe Hellebeau',2),
    ('Les Grégoire','Mme Hellebeau',2),
    ('Négrel','Philippe Hellebeau',4),
    ('Négrel','Mme Hellebeau',5),
    ('Négrel','Majordome',1),
    ('Négrel','Toussaint Maheu',1),
    ('Négrel','Etienne Lantier',1),
    ('Cécile','La maheude',2),
    ('Cécile','Philippe Hellebeau',2),
    ('Toussaint Maheu','La maheude',4),
    ('Toussaint Maheu','Etienne Lantier',4),
    ('Toussaint Maheu','Philippe Hellebeau',3),
    ('Etienne Lantier','La maheude',2),
    ('Etienne Lantier','Pluchart',3),
    ('Etienne Lantier','Souvarine',3),
    ('Deneulin','Philippe Hellebeau',2),
    ('Deneulin','Jeanne',3),
    ('Deneulin','Lucie',3),
    ('Mme Hellebeau','Philippe Hellebeau',4),
    ('Mme Hellebeau','Jeanne',2),
    ('Mme Hellebeau','Lucie',2),
    ('Mme Hellebeau','Majordome',2),
    ('Majordome','Philippe Hellebeau',1),
    ('Pluchart','Souvarine',3),
]

G = nx.Graph()
G.add_weighted_edges_from(edges)

A = nx.to_numpy_array(G) # matrice d'adjacence

n_clusters = 3
sc = SpectralClustering(n_clusters=n_clusters, affinity='precomputed', random_state=42) 
labels = sc.fit_predict(A)
```

```python
pos = nx.spring_layout(G) 
colors = [labels[node] for node in range(len(G.nodes()))]
nx.draw(G, pos, with_labels=True, node_color=colors, cmap=plt.cm.rainbow)
plt.show()
```
